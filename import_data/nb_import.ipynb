```python
df_employee = pd.read_excel('udin.lecturer.lecturer.xls')
df_employee.fillna("", inplace=True)
df_employee
```

```python
df_department = pd.read_excel('hr.department 1.xls')
df_department.fillna("", inplace=True)
df_department
```

```python
client.search("docav.department", [])
```

```python
erppeek.Model(client, "docav.department").keys()
```

```python
tipe_unit_list = [('0', 'Universitas'), ('1','Fakultas'),('2','Jurusan'), ('3','Prodi'),('4','Laboratorium'),('5','UPT'),('6','Penyelenggara MKU'),('7', 'Ormawa')]

for i, depart in df_department.iterrows():
    if(client.search('docav.department', [("nama_inggris", "=", depart["Nama Inggris"])])):
        continue
    print(depart["Nama Inggris"])
    search_tipe = [item for item in tipe_unit_list if item[1] ==  depart["Sekup Kerja"]]
    values = {
        'tipe_unit': search_tipe[0][0],
        'name': depart["Nama Inggris"],
        'nama_inggris': depart["Nama Inggris"],
        'singkatan': depart["Singkatan"]
    }
    client.create("docav.department", values)
```
