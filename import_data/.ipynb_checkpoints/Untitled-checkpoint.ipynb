```python
import pandas as pd
import erppeek
from conn import *
```

```python
Q_GET_DEPARTMENT = """SELECT distinct b.satker as scope_kerja
                    FROM simpeg.tb_01 a
                    LEFT JOIN simpeg.skpd1 b
                    ON a.idjab= b.idskpd
                    WHERE a.idstspeg IN ('1','2','3','7')
                    AND a.idjenkedudupeg='01'"""
```

```python
cur.execute(Q_GET_DEPARTMENT)
department_sql = cur.fetchall()
```

```python
department_sql
```

```python
df_department = pd.DataFrame(department_sql)
df_department
```

### MENDAPATKAN DATA PEGAWAI

```python
Q_GET_PEGAWAI = """SELECT a.staf_id, a.nip, a.gdp, a.nama, a.gdb, 
                a.idskpd, b.satker as scope_kerja,a.idjab, b.jab as struktural, a.email_dsn as email
                FROM simpeg.tb_01 a
                LEFT JOIN simpeg.skpd1 b
                ON a.idjab= b.idskpd
                WHERE a.idstspeg IN ('1','2','3','7')
                AND a.idjenkedudupeg='01'"""
```

```python
cur.execute(Q_GET_PEGAWAI)
pegawai_sql = cur.fetchall()
```

```python
df_pegawai = pd.DataFrame(pegawai_sql)
df_pegawai["email"] = df_pegawai["email"].str.strip()
df_pegawai.head()
```

```python
for i, employee in df_pegawai.iterrows():
    check = client.search("docav.pegawai", [("name","=",employee["nama"])])
    if len(check) > 0:
        continue
    # resource = client.create('resource.resource',{'name': employee["name"]})
    if employee["nama"]== None or employee["email"] == None:
        continue
    if employee["scope_kerja"] != None:
        scope_id = client.search("docav.department", [("name","=",employee["scope_kerja"])])
    values = {
        'name': employee["nama"],
        'gelar_depan': employee["gdp"],
        'gelar_belakang': employee["gdb"],
        'email': employee["email"],
        'scope_kerja': scope_id[0] if len(scope_id) > 0 else False,
        "structural": str(5)
    }
    client.create("docav.pegawai", values)
    print(values)
```
